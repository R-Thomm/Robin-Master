# Multiprocessing and joblib

```python

# print("number of CPU's:", mp.cpu_count())

def func(x):
    return(x**x)
```

```python
import multiprocessing as mp

t1 = time.time()
if __name__ == "__main__":
    pool = mp.Pool(8)
    res = pool.map(func, range(1000))

time.sleep(0.1)
print(res[1:10])
print("processing time:", np.round( time.time()-t1, 4))
```

```python
from joblib import Parallel, delayed

t1 = time.time()
res = Parallel(n_jobs=-1)(delayed(func)(2) for i in range(1000))
print(res[1:10])
print("processing time:", np.round( time.time()-t1, 4))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from joblib import Parallel, delayed

def func_plot(a):
    x = np.linspace(0,1,10)
    y = a*x**2
    plt.plot(x, y)
    plt.show()
    return x, y
```

```python
a_list = [0.5,1,1.5]

res = [func_plot(a) for a in a_list]
print(res)

res = Parallel(n_jobs=-1)(delayed(func_plot)(a) for a in a_list)
print(res)
```

# try adaptive

## 1D

```python
from time import sleep
from random import random
import adaptive

from functools import partial

offset = 0.1

def f(x, offset=offset, wait=True):

#     time.sleep(0.1)
    a = 0.01
    if wait:
        sleep(random() / 10)
    return x + a**2 / (a**2 + (x - offset)**2)

learner = adaptive.Learner1D(f, bounds=(-1, 1))
```

```python
adaptive.notebook_extension()
```

```python
runner = adaptive.Runner(learner, goal=lambda l: l.loss() < 0.01)
```

```python
runner.live_info()
runner.live_plot(update_interval=0.02)
```

```python
learner2 = adaptive.Learner1D(f, bounds=learner.bounds)

xs = np.linspace(*learner.bounds, len(learner.data))
learner2.tell_many(xs, map(partial(f, wait=False), xs))

learner.plot() + learner2.plot()
```

```python
def f(x):
    return np.sin(1/x)

learner = adaptive.Learner1D(f, bounds=(0.0001, 2))
runner = adaptive.Runner(learner, goal=lambda l: l.loss() < 0.05)
runner.live_info()
runner.live_plot(update_interval=0.02)
```

## 2D

```python
def ring(xy, wait=True):
    import numpy as np
    from time import sleep
    from random import random
    if wait:
        sleep(random()/10)
    x, y = xy
    a = 0.2
    return y + np.exp(-(x**2 + y**2 - 0.75**2)**2/a**4)

learner = adaptive.Learner2D(ring, bounds=[(-1, 1), (-1, 1)])
```

```python
runner = adaptive.Runner(learner, goal=lambda l: l.loss() < 0.01)

runner.live_info()

def plot(learner):
    plot = learner.plot(tri_alpha=0.2)
    return (plot.Image + plot.EdgePaths.I + plot).cols(3)

runner.live_plot(plotter=plot, update_interval=0.1)
```

```python
sPs = np.array(list(learner.data.keys()))
sVs = np.array(list(learner.data.values()))
n_samp = len(sPs)

plt.scatter([sPs[i][0] for i in range(n_samp)], [sPs[i][1] for i in range(n_samp)])
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib widget

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter([sPs[i][0] for i in range(n_samp)], [sPs[i][1] for i in range(n_samp)], sVs)
plt.show()
```

```python
def tableau(xy):
    x, y = xy
    if x**2 + 0.66*y**2 < 1:
        if x**2 + 0.66*y**3 < 0.4:
            return 0.5, 1
        else:
            return 1, 0.5
    else:
        return 0, 0
```

```python
learner = adaptive.Learner2D(tableau, bounds=[(-2, 2), (-2, 2)])

runner = adaptive.Runner(learner, goal=lambda l: l.loss() < 0.01)

runner.live_info()

def plot(learner):
    plot = learner.plot(tri_alpha=0.2)
    return (plot.Image + plot.EdgePaths.I + plot).cols(3)

runner.live_plot(plotter=plot, update_interval=0.1)
```

```python
sPs = np.array(list(learner.data.keys()))
sVs = np.array(list(learner.data.values()))
n_samp = len(sPs)

plt.scatter([sPs[i][0] for i in range(n_samp)], [sPs[i][1] for i in range(n_samp)])
plt.show()
```

```python
times = np.linspace(-2,1,50)
phis = np.linspace(0, 2*np.pi, 50)

xs, ys, zs = [], [], []
for t in times:
    for p in phis:
        xs.append(np.cosh(t) * np.cos(p))
        ys.append(np.cosh(t) * np.sin(p))
        zs.append(t)

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,7.5))
ax = Axes3D(fig)

ax.plot(xs, ys, zs)
plt.show()
```

```python
times = np.linspace(0.1, 0.5*np.pi-0.0001, 50)
phis = np.linspace(0, 2*np.pi, 50)

xs, ys, zs = [], [], []
for t in times:
    for p in phis:
        xs.append(np.sin(t) * np.cos(p))
        ys.append(np.sin(t) * np.sin(p))
        zs.append(np.cos(t) + np.log(np.tan(0.5*t)))

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12,7.5))
ax = Axes3D(fig)

ax.plot(xs, ys, zs)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))

ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(np.cosh(times), times)
ax1.set_xlabel(r'cosh(t)')
ax1.set_ylabel('t')

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.plot(xs, ys, zs)
plt.show()
```
