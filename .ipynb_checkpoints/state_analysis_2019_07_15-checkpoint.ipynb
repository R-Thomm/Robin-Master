```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import pickle
import os.path
import json

import sys
sys.path.insert(0, './../')

import PyModules.analyse_eios.eios_data as eios_file
import PyModules.analyse_eios.eios_analyse as eios_analyse
import PyModules.analyse_eios.eios_sb as eios_sb
import PyModules.pdq.waveform as waveform
from PyModules.analyse_eios.eios_data import find_files

path_data_local = '/home/qsim/Results/tiamo4.sync'
path_data_remote = '/mnt/qsim/tiamo4/messungen/EIOS'
cache_path='./data/'

def print_ips(name, ips):
    _, root = eios_file.read(name)
    ip = eios_file.read_xml(root)['ionproperties']
    #print('File:', name)
    print('----------------')
    print('Ion properties:')
    print('----------------')
    ip_l = []
    for i in ips:
        print(i+':', ip[i])
        ip_l.append(ip[i])
    print('----------------')    
    return ip_l

def ana_mf_sq(name, fix=[0,0,0,0,1,0,0], nmax = 15, tangle = 38.0):
    redflop, blueflop, lbl = eios_sb.open_file([name], cache_path)
    ip_l = print_ips(name, ['wf_id', 'amp', 't_wbtw', 'Damp2', 'fr_mf', 'rabi_mf', 'n_th_mf']) 
    f_mf = ip_l[4] 
    LD = eios_sb.LDparameter(f_mf,tangle)
    Rabi_init = ip_l[5]
    dec_init = 0.01
    limb_init = 0.5
    limr_init = 0.9
    nth = ip_l[6]
    ncoh = 0.0001
    nsq = 0.0001
    inits = [Rabi_init,dec_init,limb_init,limr_init, nth, ncoh, nsq]

    red_chi, fmin, param = eios_sb.fit_RSB_BSB_par(redflop, blueflop, LD, nmax, inits, fix, ntrot = 1, doplot = True)
    return ip_l, red_chi, fmin, param

def ana_mf_sq_mod(name, fix=[0,0,0,0,1,0,0], nmax = 15, tangle = 38.0):
    redflop, blueflop, lbl = eios_sb.open_file([name], cache_path)
    ip_l = print_ips(name, ['EU_beam_conf','t_tickle_mf_EU', 'u_tickle_mf_EU', 'fr_mf_sq', 'fr_mf', 'rabi_mf', 'n_th_mf', 'EU_t_mf_1R']) 
    f_mf = ip_l[4] 
    LD = eios_sb.LDparameter(f_mf,tangle)
    Rabi_init = ip_l[5]
    dec_init = 0.01
    limb_init = 0.5
    limr_init = 0.9
    nth = ip_l[6]
    ncoh = 0.0001
    nsq = 0.0001
    inits = [Rabi_init,dec_init,limb_init,limr_init, nth, ncoh, nsq]

    red_chi, fmin, param = eios_sb.fit_RSB_BSB_par(redflop, blueflop, LD, nmax, inits, fix, ntrot = 1, doplot = True)
    return ip_l, red_chi, fmin, param
```
