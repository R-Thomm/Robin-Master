```python
import numpy as np
import scipy.optimize as sco
import matplotlib.pyplot as plt
```

```python
def par_fit(x,a,b,c):
    return -a*x**2 + b*x + c
```

## Shim Coils (Magnet): NS

```python
shim_ns_c =  [1.35, 1.20, 1.05, 0.9, 0.8, 0.7, 0.6, 0.0, 3.0, 4.0, 5.0, 6.0, 0, 2.0, 1.4]
fidelity_b = [95, 95, 93, 93, 95, 94, 94, 93, 93, 92, 88, 86, 90, 95, 94]
fidelity_d = [95, 94, 95, 94, 93, 95, 94, 94, 94, 92, 93, 89, 93, 94, 94]
fr_3p3_2p2 = [1775.0812, 1775.0323, 1774.974,  1774.927,  1774.895,  1774.856,  1774.8214, 1774.570,  1775.569,  1775.8243, 1776.087,  1776.372 , 1774.5761, 1775.302 , 1775.117 ]
fr_3p1_2p2 = [1780.5969, 1780.5694, 1780.5327, 1780.5064, 1780.4864, 1780.4651, 1780.4423, 1780.2934, 1780.8871, 1781.0395, 1781.1936, 1781.3636, 1780.2968, 1780.7293, 1780.6180]
```

```python
plt.plot(shim_ns_c,fr_3p3_2p2,'kx')
plt.title('Shim Coils (Magnet): NS, 3p3p-2p2')
plt.ylabel('frequency of 3p3-2p2 transition (MHz)')
plt.xlabel('current threw magnet coils (arb.u.)')
plt.show()

plt.plot(shim_ns_c,fr_3p1_2p2,'kx')
plt.title('Shim Coils (Magnet): NS, 3p1-2p2')
plt.ylabel('frequency of 3p1-2p2 transition (MHz)')
plt.xlabel('current threw magnet coils (arb.u.)')
plt.show()
```

```python
y=min(fidelity_b,fidelity_d)
plt.errorbar(shim_ns_c,y,1, fmt='x')
min(fidelity_b,fidelity_d)
popt, pcov = sco.curve_fit(par_fit, shim_ns_c, y)
z = np.linspace(min(shim_ns_c),max(shim_ns_c),100)
plt.plot(z, par_fit(z,*popt))
print('maximum:', -popt[1]/(-2*popt[0]))
plt.title('Shim Coils (Magnet): NS, fidelity')
plt.ylabel('fidelity (%)')
plt.xlabel('current threw magnet coils (arb.u.)')
plt.show()
```
